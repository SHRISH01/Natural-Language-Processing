            <p style="font-size: 1.2rem;">
                <strong>Topic Modeling</strong> is a type of statistical modeling used to discover hidden semantic structures in a collection of documents. It allows you to automatically identify topics within large amounts of text data. This technique is crucial for organizing, understanding, and summarizing vast collections of information.

            <h2 style="; font-size: 2 rem; font-weight: bold;">Why BERTopic is Special:</h2>
            <ul style="font-size: 1.1rem;">
                <li><strong>Transformer-Based Embeddings</strong> - BERTTopic uses BERT or other transformer models to create high-quality, contextualized embeddings of documents. Unlike traditional bag-of-words approaches, BERT embeddings capture semantic meanings and context, resulting in more accurate topic clusters.</li>
                <li><strong>Dynamic Topic Modeling</strong> - BERTTopic can dynamically group similar topics together over time, which is useful for analyzing trends and topic changes in evolving text data.</li>
                <li><strong>Ease of Interpretation</strong> - BERTTopic generates interpretable topics by reducing dimensionality with techniques like UMAP (Uniform Manifold Approximation and Projection) and clustering embeddings using HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise).</li>
                <li><strong>Interactive Visualizations</strong> - It also offers interactive tools for visualizing topics, topic distributions, and relations, making it easier for analysts to understand and interpret the model outputs.</li>

```python
# remove the big comments
df = df[df['word_length'] <= 2000]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="word_length", hue="Rating", multiple="stack", palette="bright", bins=200)
plt.title('Distribution of Text Lengths by Class', fontsize=16)
plt.xlabel('Text Length', fontsize=14)
plt.ylabel('Frequency')
plt.show()
```

```python
# clean the text

def clean_text(text):

  text = text.lower()

  text = re.sub(r'([^\w\s])', r' \1 ', text)

  text = re.sub(r'\s+', ' ', text, flags=re.I)

  words = word_tokenize(text)

  text = ' '.join(w for w in words if len(w) > 1)

  return text.strip()

df['Review'] = df['Review'].apply(clean_text)
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    🔗 Embedding
  </h2>
</div>

```python
model_id = "sentence-transformers/distiluse-base-multilingual-cased-v2"

embedding_model = SentenceTransformer(model_id, device='cuda:0')
```

```python
reviews_embeddings = embedding_model.encode(df['Review'].values, show_progress_bar=True)
```

```python
reviews_embeddings.shape
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    ✂ Demintional Reduction
  </h2>
</div>

```python
umap_model = UMAP(n_neighbors=15,
                  n_components=15,
                  min_dist=0.0,
                  metric='cosine',
                  random_state=42)
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    🕸️ Clustring
  </h2>
</div>

```python
hdbscan_model = HDBSCAN(min_cluster_size=50,
                        metric='euclidean',
                        cluster_selection_method='eom',
                        prediction_data=True)
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    🔢 Vectorizer
  </h2>
</div>

```python
vectorizer_model = CountVectorizer(
    stop_words='english',
    min_df=3, 
    analyzer='word',
    max_df=0.5,
    ngram_range=(1, 3)
)
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    👨🏼‍💻 Topic Represinter
  </h2>
</div>

```python
keybert_model = KeyBERTInspired()

representation_model = {
    "keyBERT": keybert_model
}
```

<div style="font-family: Arial, Helvetica, sans-serif; background-color: #2c3e50; color: white; padding: 20px; border-radius: 8px; text-align: center; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">
  <h2 style="font-size: 2rem; font-weight: bold; margin: 0;">
    📐 Model Building
  </h2>
</div>

```python
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    representation_model=representation_model,

    top_n_words=10,
    verbose=True
)

topics, probs = topic_model.fit_transform(df['Review'].values, reviews_embeddings)
```

```python
topic_model.get_topic_info()
```
